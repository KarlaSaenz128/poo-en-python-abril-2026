```python
# Definimos la clase Animal
class Animal:
    # El método __init__ es el constructor (donde creamos los atributos)
    # self hace referencia al objeto especifico que estamos creando
    def __init__(self, nombre, raza):
        self.nombre = nombre
        self.raza = raza

    # Función para que el animal ladre
    def ladrar(self):
        print(f"¡Guau! El perro {self.nombre} dice: !Guau,guau.")

    # Función para que el animal corra
    def correr(self):
        print(f"El {self.nombre} lamado {self.nombre} está corriendo velozmente .")


# ---- creacion de los objetos (instancion)--

# Creamos el objeto Pitbull
pitbull = Animal("Thor", "Pitbull")


# Creamos el objeto Chihuahua con datos especificos
chihuahua = Animal("Chispa", "Chihuahua")
# creamos el objeto perro callejero con datos
perrocallejero = Animal("solovino", "chafa")

# --- Usar las funciones ---

print(f" datos del primer perro: {pitbull.nombre}, {pitbull.raza}")
pitbull.ladrar()
pitbull.correr()

print(f" Datos del segundo perro: {chihuahua.nombre}, Raza: {chihuahua.raza}.")
chihuahua.ladrar()
chihuahua.correr()
print("-"* 30) # linea divisora
print(f" Datos del tercer perro: {perrocallejero.nombre}, Raza: {perrocallejero.raza}.")
perrocallejero.ladrar()
perrocallejero.correr()
print("elaborado por Karla Saenz 0133")
```
