```python
# Definición de la clase Cantante
class Cantante:
    # El método __init__ inicializa los 3 atributos
    def __init__(self, nombre, edad, genero):
        self.nombre = nombre
        self.edad = edad
        self.genero = genero

    # Función 1: componer
    def componer(self):
        print(f"{self.nombre} Adelen {self.genero} pop.")

    # Función 2: funcion (el evento/show)
    def funcion(self):
        print(f"La función ha comenzado: ¡{self.nombre} Adelen!")

    # Función 3: cantar
    def cantar(self):
        print(f"{self.nombre} está cantando a sus {self.edad} 35 años.")

# --- Creación de los 2 objetos (Instancias) ---

# Objeto cantante1 con datos de ejemplo
cantante1 = Cantante("Adele", 35, "Soul")

# Objeto cantante2 con datos de ejemplo
cantante2 = Cantante("Bruno Mars", 38, "Funk")

# --- Ejecución de las funciones para ver los mensajes ---

print("--- ACTIVIDADES DE CANTANTE 1 ---")
print(f"Artista: {cantante1.nombre}")
cantante1.componer()
cantante1.funcion()
cantante1.cantar()

print("\n--- ACTIVIDADES DE CANTANTE 2 ---")
print(f"Artista: {cantante2.nombre}")
cantante2.componer()
cantante2.funcion()
cantante2.cantar()

print("Elaborado por Karla Saenz 0133 ")
```
